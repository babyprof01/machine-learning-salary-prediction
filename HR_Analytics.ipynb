```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load dataset
data = pd.read_csv("/content/HR_dataset.csv")

# Quick look at the data
print(data.head())
print(data.info())
```

```python
# Check total rows and duplicates
print("Total rows:", len(data))
duplicates = data.duplicated()
print("Number of duplicate rows:", duplicates.sum())
```

```python
data[data.duplicated(keep=False)].head(10)
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
#Retention distribution
data['left'].value_counts(normalize=True)
```

```python
#correlation of numerical variables
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
#compare means by correlation
data.groupby("left").mean(numeric_only=True)
```

```python
#bar chart of salary vs retention
salary_retention = data.groupby("salary")["left"].mean()

plt.figure(figsize=(6,4))
salary_retention.plot(kind="bar")
plt.title("Salary Level vs Leaving Rate")
plt.ylabel("Proportion Leaving")
plt.show()
```

```python
#department vs retention. This shows which departments experience higher attrition.
dept_retention = data.groupby("Department")["left"].mean()

plt.figure(figsize=(10,5))
dept_retention.sort_values().plot(kind="bar")
plt.title("Department vs Leaving Rate")
plt.ylabel("Proportion Leaving")
plt.xticks(rotation=45)
plt.show()
```

Build Logistic Regression Model

```python
#encode categorical variables
data_encoded = pd.get_dummies(data, columns=["salary", "Department"], drop_first=True)
```

```python
#define x and y variables
X = data_encoded.drop("left", axis=1)
y = data_encoded["left"]
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
#Logistic regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

Measure model accuracy

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_prob = model.predict_proba(X_test)[:,1]
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

The logistic regression model achieved an accuracy of 82.9% and an AUC of 0.831, indicating good discriminatory ability. This suggests that employee satisfaction, workload, tenure, promotion status, and salary level are strong predictors of attrition.
